**Exercise 1:** Find the solution of the linear systems

```python
import sympy as sp

x, y, z, t = sp.symbols("x, y, z, t")

# cau a
equation_1 = sp.Eq(x + 2*y + z, 0)
equation_2 = sp.Eq(2*x - y + z, 0)
equation_3 = sp.Eq(2*x + y, 0)

resolution = sp.solve((equation_1, equation_2, equation_3), (x, y, z))
print("Cau a:", resolution)

# cau b
equation_1 = sp.Eq(2*x + y + z + t, 1)
equation_2 = sp.Eq(x + 2*y + z + t, 1)
equation_3 = sp.Eq(x + y + 2*z + 2*t, 1)
equation_4 = sp.Eq(x + y + z + 2*t, 1)
resolution = sp.solve((equation_1, equation_2, equation_3, equation_4), (x, y, z, t))
print("Cau b:", resolution)
```

```python
# Another way
import numpy as np

# Bên trái dấu "=" là ma trận A
# Bên phải dâu "=" la ma trận B
# .T để chuyển vị ma trận (Transpose)

A_matrix = np.array([
    [1, 2, 1],
    [2, -1, 1],
    [2, 1, 0]
])
B_matrix = np.array([0, 0, 0]).T
resolution = np.linalg.solve(A_matrix, B_matrix) 
print(resolution)

```

**Exercise 2:** Consider the following system of linear equations:

a1x + b1y = c1

a2x + b2y = c2

a3x + b3y = c3

For real values of ai, bi and ci, the graphs of these equations are lines in a two dimensional (x, y) coordinate system. Write a function to investigate the number of solutions for system in two variables
such that

(a) The system has no solutions.

(b) The system has a unique solution.

(c) The system has infinitely many solutions.

Trong đoạn code này, chúng ta đang xem xét phương trình đường thẳng hai chiều dạng:

ax + by = c

Nếu ta giải phương trình này, ta sẽ được phương trình đường thẳng dạng:

y = (-a/b)x + c/b

Trong đó, a, b, và c là các hằng số và x và y là các biến.
Dựa trên định nghĩa này, chúng ta có thể xác định được số lượng nghiệm của hai đường thẳng bằng cách xem xét giá trị của hệ số góc a/b.

Nếu a/b giống nhau cho cả hai đường thẳng, thì chúng sẽ là hai đường thẳng song song và không giao nhau. Trong trường hợp này, hệ phương trình không có nghiệm. Đây là trường hợp được mô tả bởi code đầu tiên.

Nếu a/b giống nhau và c/b khác nhau, thì chúng ta có hai đường thẳng trùng nhau. Trong trường hợp này, hệ phương trình có vô số nghiệm. Đây là trường hợp được mô tả bởi code thứ hai.

Nếu a/b khác nhau cho cả hai đường thẳng, thì chúng ta có hai đường thẳng giao nhau tại một điểm duy nhất. Trong trường hợp này, hệ phương trình có một nghiệm duy nhất. Đây là trường hợp được mô tả bởi code cuối cùng.



```python
import numpy as np
import matplotlib.pyplot as plt
import random



def plot_2d_equation(x_range, a, b, c):
  # ax + by = c, where x is in x_range -> y = (c - ax) / b
  if b != 0:
    fx = lambda x: (c - a * x) / b
    y_range = list(map(fx, x_range))
    plt.plot(x_range, y_range)
  else:
    if a != 0:
      x_range_new = np.full(len(x_range), c / a)
      print("new x range =", x_range_new)
      y_range = np.linspace(-10, 10, num=len(x_range))
      plt.plot(x_range_new, y_range, label=str(a) + " + x = " + str(c))
    else:
      print("Cannot plot this function")

def randomNumber():
  return random.randint(1, 9)

start_value = -10
stop_value = 10
step_value = 0.1

x_range = np.arange(start_value, stop_value + step_value, step_value)

# Đường thẳng song song - Hệ số góc bằng nhau -> y = -(a/b)*x + c/b
# Không có nghiệm: 2 đường thẳng song song nhau không giao nhau.
plt.title("No solution")
plot_2d_equation(x_range, -1, 1, 3)
plot_2d_equation(x_range, 3, -3, 6)

# Đường thẳng trùng nhau: a1, b1, c1 trùng (a1, b1, c1)*k
# Vô số nghiệm: 2 đường thẳng trùng nhau, tương ứng với 2 phương trình đường thẳng bằng nhau nhân với một hằng số k.

# Tạo ra các hệ số ngẫu nhiên cho phương trình đường thẳng ban đầu
a1, b1, c1 = randomNumber(), randomNumber(), randomNumber()

# Tạo ra một số nguyên ngẫu nhiên để tạo ra hằng số k
k = random.randint(1, 9)

# Tạo ra các hệ số cho phương trình đường thẳng mới bằng cách nhân các hệ số của phương trình đường thẳng ban đầu với hằng số k
a2, b2, c2 = a1 * k, b1 * k, c1 * k

plt.show()
plt.title("Infinitely solution")
plot_2d_equation(x_range, a1, b1, c1)
plot_2d_equation(x_range, a2, b2, c2)

# Đường thẳng giao nhau:
# Một nghiệm duy nhất: 2 đường thẳng giao nhau tại một điểm duy nhất trên mặt phẳng.
a, b, c = randomNumber(), randomNumber(), randomNumber()
d, e, f = randomNumber(), randomNumber(), randomNumber()

plt.show()
plt.title("Unique solution")
plot_2d_equation(x_range, a, b, c)
plot_2d_equation(x_range, d, e, f)
```

**Exercise 3:** Consider the following system of linear equations:

a1x + b1y + c1z = d1

a2x + b2y + c2z = d2

a3x + b3y + c3z = d3

For real values of ai, bi, ci and di, the graphs of these equations are the planes in a three dimensional (x, y, z) coordinate system. You should write a function to investigate the number of solutions for system in two variables such that

(a) The system has no solutions.

(b) The system has a unique solution.

(c) The system has infinitely many solutions.

```python
import numpy as np
import matplotlib.pyplot as plt 

def plot3DEquation(ax, x_ar, a, b, c, d):
    """
    Plots a 3D surface for the equation ax + by + cz = d.
    
    Parameters:
    ax (Axes3D): The 3D axes object.
    x_ar (array): The range of x values.
    a (float): The coefficient of x.
    b (float): The coefficient of y.
    c (float): The coefficient of z.
    d (float): The constant term.
    """
    if c != 0:
        # z = (d - ax - by) / c
        z_ar = (d - a*x_ar[:,None] - b*x_ar[None,:]) / c
        X, Y = np.meshgrid(x_ar, x_ar)
        ax.plot_surface(X, Y, z_ar)
    else:
        if b != 0:        
            # y = (d - ax) / b
            y_ar = (d - a*x_ar) / b
            ax.plot(x_ar, y_ar, zs=-50)
        else:
            if a != 0:
                # x = d / a
                x_val = d / a
                y_ar = np.linspace(-10, 10, len(x_ar))
                ax.plot([x_val]*len(y_ar), y_ar, zs=-50)
            else:
                print("Cannot plot this function!")

# test the function
x_ar = np.arange(-10, 10.1, 0.1)
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
plot3DEquation(ax, x_ar, 5, 1, 6, 10)
plot3DEquation(ax, x_ar, 2, 12, 8, 20)
plot3DEquation(ax, x_ar, 3, 9, 3, 1) 

plt.title("System 2")
ax.set_xlim([-10, 10])
ax.set_ylim([-20, 20])
ax.set_zlim([-50, 50])
plt.show()
```

Exercise 5: Try to use the three different methods to solve the following system:

x + 2y+z=1

2x+ 2y+2z=1

2x+ 4y+z=2

```python
#bai 5

import numpy as np
import sympy as sp 

#cach 1:

x, y, z = sp.symbols('x, y, z')

eq1 = sp.Eq(x + 2*y + z, 1)
eq2 = sp.Eq(2*x + 2*y + 2*z, 1)
eq3 = sp.Eq(2*x + 4*y + z, 2)

sol = sp.solve((eq1, eq2, eq3), (x, y, z))
print("cach 1:\n", sol)

#cach 2:

A = np.array([
    [1, 2, 1],
    [2, 2, 2],
    [2, 4, 1]
])

b = np.array([[1], [1], [2]])

sol1 = np.linalg.solve(A, b)
print("cach 2:\n", sol1)

#cach 3 - su dung cong thuc tuyen tinh - invert A * b 

#tim ma tran nghich dao cua ma tran A bang - np.linalg.inv()
A_inv = np.linalg.inv(A)

sol2 = np.matmul(A_inv, b)

print("cach 3:\n", sol2)
```

Exercise 6: Find the interpolating polynomial p(t) = a0+a1t+a2t
2 for the data (1,6), (2,15), (3,38). That is, find a0, a1 and a2

```python
#bai 6 

import sympy as sp

a0, a1, a2 = sp.symbols('a0, a1, a2')

eq1 = sp.Eq(a0 + a1 + a2, 6)              #t = 1, pt = 6
eq2 = sp.Eq(a0 + a1*2 + a2*2**2, 15)      #t = 2, pt = 15
eq3 = sp.Eq(a0 + a1*3 + a1*3**2, 38)      #t = 3, pt = 38 

solve = sp.solve((eq1, eq2, eq3), (a0, a1, a2))
print(solve)
```

Exercise 7: A group took a trip on a bus, at $3 per child and $3.2 per adult for a total of $118.4. They
took the train back at $3.5 per child and $3.6 per adult for a total of $135.2. How many children and
how many adults?

```python
#bai 7

import sympy as sp

child, adult = sp.symbols('child, adult')

eq1 = sp.Eq(3*child + 3.2*adult, 118.4)
eq2 = sp.Eq(3.5*child + 3.6*adult, 135.2)

solve = sp.solve((eq1, eq2), (child, adult))
print(solve)
```

Exercise 14: Save even integer numbers in the matrix A into a vector.

```python
#bai 14

import numpy as np

A = np.random.randint(1, 100, size = (10, 10))

B = []

for i in range(10):
  for j in range(10):
    if A[i][j] % 2 == 0:
      B.append(A[i][j])

print("Matrix A = \n", A)
print("Even integer in the matrix A = \n", B)
```
